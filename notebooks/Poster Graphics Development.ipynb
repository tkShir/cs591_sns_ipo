```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

# Graphics for Valuation Comparison

```python
valuation_dict = {
    "Palantir" : 13615104822,
    "ZoomInfo" : 8175471698,
    "Unity" : 14080376308,
    "Asana" : 3246679245,
    "Snowflake" : 33972040237,
    "Kingsoft Cloud" : 3706302521
}

chart_title = "Market Capitalization derived by IPO Reference Price"
x_axis_title = "Company Name"
y_axis_title = "Market Capitalization (USD)"

x_pos = [i for i in range(len(valuation_dict))]

company_name, valuation_data = list(), list()
for key, val in valuation_dict.items():
    company_name.append(key)
    valuation_data.append(val)

plt.figure(figsize=(20,10))
plt.bar(x_pos, valuation_data, color='blue')
plt.xlabel(x_axis_title)
plt.ylabel(y_axis_title)
plt.title(chart_title)


plt.xticks(x_pos, company_name)

plt.show()
```
